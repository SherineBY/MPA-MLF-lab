# SVM for classification

Load IRIS dataset, check its contents:


```python
from sklearn.datasets import load_iris
iris=load_iris()
iris.feature_names
print(iris.feature_names)
print(iris.data[:,:])
print(iris.target)
#print(iris.data)
```

Split data into training and testing parts:

```python
from sklearn.model_selection import train_test_split
X=iris.data
y=iris.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(X_train.shape)
print(X_test.shape)
```

Use a Support Vector Machine for classification:


```python
from sklearn.svm import SVC
SVMmodel=SVC(kernel='linear')
SVMmodel.fit(X_train,y_train)
SVMmodel.get_params()
SVMmodel.score(X_test,y_test)
```

Let's explore more now.

*   Choose only first two features (columns) of iris.data
*   SVM is in its basic form a 2-class classifier, so eliminate iris.target =2 from the data



```python
import numpy as np

# Prendre uniquement les deux premières colonnes
X_2D = iris.data[:, :2]
y = iris.target

# Supprimer la classe 2 pour un problème binaire (classes 0 et 1 uniquement)
mask = y != 2  # Garde seulement les éléments où y est différent de 2
X_2D = X_2D[mask]
y_binary = y[mask]

print(f"Nouvelle taille des données : {X_2D.shape}")
print(f"Classes restantes : {np.unique(y_binary)}")  # Vérifier qu'on a bien 0 et 1
```

Plot scatterplots of targets 0 and 1 and check the separability of the classes:

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
plt.scatter(X[y==0,0],X[y==0,1], color='blue')
plt.scatter(X[y==1,0],X[y==1,1], color='red')
plt.scatter(X[y==2,0],X[y==2,1], color='cyan')
```

Train and test the SVM classifier, play with regularization parameter C (either use the default value or try e.g. 200)

```python
from sklearn.svm import SVC

X= iris.data[iris.target!=2,0:2]
print (X.shape)
y=iris.target[iris.target!=2]
SVMmodel_1=SVC(kernel='linear',C=200)

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X,y,test_size=0.2)
SVMmodel_1.fit(X_train_1,y_train_1)
#print (SMmodel_1-get_params ())
SVMmodel_1.score(X_test_1,y_test_1)
```

Show support vectors in the 2D plot, plot the decision line from equation [w0 w1]*[x0 x1] + b = 0:

```python
supvectors=SVMmodel_1.support_vectors_
# Plot the support vectors here


#Separating line coefficients:
W=SVMmodel_1.coef_
b=SVMmodel_1.intercept_
print(W)
print(b)
import numpy as np
x1=np.linspace(np.min(X[:,0]),np.max(X[:,0]))
x2=-b/W[0,1]-W[0,0]/W[0,1]*x1 # Use the correct model (SVMmodel_1) for coefficients and intercept
plt.scatter(X[y==0,0], X[y==0,1],color='blue')
plt.scatter(X[y==1,0],X[y==1,1], color='red')
plt.scatter(x1,x2, color='black')
plt.show() # Add plt.show() to display the plot
```

# Anomaly detection via SVM

Import one-class SVM and generate data (Gaussian blobs in 2D-plane):

```python
from sklearn.svm import OneClassSVM
from sklearn.datasets import make_blobs
from numpy import quantile, where, random

random.seed(11)
x, _ = make_blobs(n_samples=300, centers=1, cluster_std=.3, center_box=(4, 4))

plt.scatter(x[:,0], x[:,1])
plt.show()

```

Train one-class SVM and plot the outliers (outputs of prediction being equal to -1)

```python
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)


SVMmodelOne.fit(x)
pred = SVMmodelOne.predict(x)
anom_index = where(pred==-1)
values = x[anom_index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()

```

Plot the support vectors:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from numpy import random
from sklearn.svm import OneClassSVM # Import OneClassSVM here

random.seed(11)
x, _ = make_blobs(n_samples=300, centers=1, cluster_std=.3, center_box=(4, 4))

# Define and fit SVMmodelOne before using it
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)
SVMmodelOne.fit(x)

plt.scatter(x[:, 0], x[:, 1], label="Normal Data")
plt.scatter(SVMmodelOne.support_vectors_[:, 0], SVMmodelOne.support_vectors_[:, 1],
            color='green', marker='x', label="Support Vectors")
plt.axis('equal')
plt.legend()
plt.title("Support Vectors Visualization")
plt.show()
```

What if we want to have a control what is outlier? Use e.g. 5% "quantile" to mark the outliers. Every point with lower score than threshold will be an outlier.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from numpy import random, quantile, where # Import where from numpy
from sklearn.svm import OneClassSVM # Import OneClassSVM here


# Define and fit SVMmodelOne before using it
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)
SVMmodelOne.fit(x)

scores = SVMmodelOne.score_samples(x)

thresh = quantile(scores, 0.5) # Now quantile is defined
print(thresh)
index = where(scores<=thresh)
values = x[index]

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()
```
